```python
import os
import sys
from pathlib import Path

from zeroheliumkit.fem.freefemer import EDPpreparer
```

```python
path = Path("zhktest")
os.makedirs(path / Path("geo"), exist_ok=True)

yaml_no_adapt = """
savedir: zhktest
meshfile: zhktest/geo/dot.msh
physicalSurfaces:
   mid: 5
   out: 6
   top: 7
physicalVolumes:
   VACUUM: 1
   DIELECTRIC: 2
   METAL: 3
   HELIUM: 4
dielectric_constants:
   VACUUM: 1.0
   DIELECTRIC: 11.0
   METAL: 1.0
   HELIUM: 1.057
ff_polynomial: 2
extract_opt:
  - name: result1
    quantity: phi
    plane: xy
    coordinate1: [-10, 10, 201]
    coordinate2: [-10, 10, 201]
    coordinate3: [1.6]
    curvature_config: null
"""

with open(path / Path("dot.yaml"), "w") as f:
    f.write(yaml_no_adapt)
```

```python
# Without adaptation:

edps = EDPpreparer(path / Path("dot.yaml"))
#with open(path + "/edp/ff_mid.edp", "r") as f:
    #print(f.read())
```

```python
# With mesh adaptation:
#os.makedirs("/tmp/zhktest2/geo", exist_ok=True)

yaml_adapt = """
savedir: zhktest
meshfile: zhktest/geo/dot.msh
physicalSurfaces:
   mid: 5
   out: 6
   top: 7
physicalVolumes:
   VACUUM: 1
   DIELECTRIC: 2
   METAL: 3
   HELIUM: 4
dielectric_constants:
   VACUUM: 1.0
   DIELECTRIC: 11.0
   METAL: 1.0
   HELIUM: 1.057
ff_polynomial: 2
adaptation_config:
  mesh_adaptation: true
  anisotropic_adaptation: false
  n_adapt: 3
  err_target: 0.01
extract_opt:
  - name: result1
    quantity: phi
    plane: xy
    coordinate1: [-10, 10, 201]
    coordinate2: [-10, 10, 201]
    coordinate3: [1.6]
    curvature_config: null
"""

with open(path / Path("dot.yaml"), "w") as f:
    f.write(yaml_adapt)

edps2 = EDPpreparer(path / Path("dot.yaml"))
#with open("/tmp/zhktest2/edp/ff_mid.edp", "r") as f:
    #print(f.read())
```
